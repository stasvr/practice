```python
import tensorflow as tf
import numpy as np
from keras.datasets import mnist
```

```python
data = mnist.load_data()
```

```python
"""
RBM
"""
class Model:
    
    # in - visible
    # out - hidden
    def __init__(self, n_visible=784, n_hidden=500):
        tf.reset_default_graph()
        self.alpha = 1.0
        
        self.n_visible = n_visible
        self.n_hidden = n_hidden
       
        self.in_tensor = tf.placeholder("float", [None, 784])
        self.out_tensor = tf.placeholder("float", [None, 500])
        
        self.W = tf.placeholder("float", [784, 500])
        self.in_b = tf.placeholder("float", [784])
        self.out_b = tf.placeholder("float", [500])       
        
        self.__build()
        
        
    def __build(self):
        alpha = self.alpha
        out_tensor_prob_start = tf.nn.sigmoid(tf.matmul(self.in_tensor, self.W) + self.out_b)
        out_tensor_start = self.sample_probability(out_tensor_prob_start)
        
        in_tensor_next = tf.nn.sigmoid(
            tf.matmul(out_tensor_prob_start, tf.transpose(self.W)) + self.in_b 
        )
        out_tensor_next = tf.nn.sigmoid(
            tf.matmul(in_tensor_next, self.W) + self.out_b
        )
        

        w_grad_p = tf.matmul(tf.transpose(self.in_tensor), out_tensor_start)
        w_grad_n = tf.matmul(tf.transpose(in_tensor_next), out_tensor_next)
        
        self.update_w = self.W + alpha * (w_grad_p - w_grad_n) / tf.to_float(tf.shape(self.in_tensor)[0])
        self.update_in_tensor = self.in_b + alpha * tf.reduce_mean(self.in_tensor - in_tensor_next, 0)
        self.update_out_tensor = self.out_b + alpha * tf.reduce_mean(out_tensor_prob_start - out_tensor_next, 0)
        
        # sampling functions
        self.out_sample = tf.nn.sigmoid(tf.matmul(self.in_tensor, self.W) + self.out_b)
        self.in_sample = tf.nn.sigmoid(tf.matmul(self.out_sample, tf.transpose(self.W)) + self.in_b)

        # cost
        self.error_sum = tf.reduce_mean(tf.square(self.in_tensor - self.in_sample))
        
    def sample_probability(self, probs):
        return tf.nn.relu(tf.sign(probs - tf.random_uniform(tf.shape(probs))))
        
"""
Estimation
"""
class Process:
    
    def __init__(self):
        1
        
    def run(self):
        1
```

```python
RBM = Model()
```

```python
sess = tf.Session()
init = tf.initialize_all_variables()
sess.run(init)
```

```python
# n - new
# o - old
n_w = np.zeros([784, 500], np.float32)
n_in_b = np.zeros([784], np.float32)
n_out_b = np.zeros([500], np.float32)
o_w = np.zeros([784, 500], np.float32)
o_in_b = np.zeros([784], np.float32)
o_out_b = np.zeros([500], np.float32)
```

```python
print(
    sess.run(RBM.error_sum, feed_dict={
        RBM.in_tensor : np.reshape(data[0][0], newshape=[60000, 784]),
        RBM.W : o_w, RBM.in_b: o_in_b, RBM.out_b : o_out_b
    })
)
```

```python
from sklearn.preprocessing import MinMaxScaler
trX = MinMaxScaler().fit_transform(np.reshape(data[0][0], newshape=[len(data[0][0]), 784]))
```

```python
batchsize = 100
gen_images = list()

for i in range(2):
    for start, end in zip(range(0, len(trX), batchsize), range(batchsize, len(trX), batchsize)):
        batch = trX[start:end]
    
        FEED = {RBM.in_tensor: batch, RBM.W : o_w, RBM.in_b: o_in_b, RBM.out_b : o_out_b}
    
        n_w = sess.run(RBM.update_w, feed_dict=FEED)
        n_in_b = sess.run(RBM.update_in_tensor, feed_dict=FEED)
        n_out_b = sess.run(RBM.update_out_tensor, feed_dict=FEED)
    
        o_w = n_w
        o_in_b = n_in_b
        o_out_b = n_out_b
    
        if start % 10000 == 0:
            print(
                sess.run(RBM.error_sum, feed_dict={
                    RBM.in_tensor : trX,
                    RBM.W : n_w, RBM.in_b: n_in_b, RBM.out_b : n_out_b
                })
            )
            gen_images.append(n_w.T)
```

```python
gen_images = gen_images[-1:][0]
```
